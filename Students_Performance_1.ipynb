```python
# Create a variable to hold the values to be ploted
to_plot = []
# This contains main dataset indexing and plot dataset creation in a single line
to_plot.append(Box( y=data_all['math score'], name="Math", showlegend=True ) )
to_plot.append(Box( y=data_all['reading score'], name="Reading", showlegend=True ) )
to_plot.append(Box( y=data_all['writing score'], name="Writing", showlegend=True ) )

# Layout, descriptions of the plot's style
layout = Layout(
    title="Total exams scores",
    xaxis=dict(
        title="Exam type",
        titlefont=dict(
            size=18,
            color='#7f0000'
        )
    ),
    yaxis=dict(
        title="Score",
        titlefont=dict(
            size=18,
            color='#007f7f'
        )
    ),
)

# This is where the plotting happens
# We call an offline plot module from the plotly library
# We tell it, that data to be plotted is contained in the to_plot variable and 
# the style of the plot is described in the layout variable

plotly.offline.iplot({
    "data": to_plot,
    "layout": layout
})
```

```python
# Choosing the rows where lunch type is "standard", creating subsets for all exams separately
lunch_st_rows = data_all["lunch"] == "standard"
lunch_st_math = data_all[lunch_st_rows]['math score']
lunch_st_read = data_all[lunch_st_rows]['reading score']
lunch_st_write = data_all[lunch_st_rows]['writing score']

# Choosing the rows where lunch type is "standard", creating subsets for all exams separately
lunch_red_rows = data_all["lunch"] == "free/reduced" 
lunch_red_math = data_all[lunch_red_rows]['math score']
lunch_red_read = data_all[lunch_red_rows]['reading score']
lunch_red_write = data_all[lunch_red_rows]['writing score']

# Calculate the statistical significance for each exam based on student's lunch type
(t, p) = scipy.stats.ttest_ind(lunch_st_math, lunch_red_math)
print("P-Value for the Math score: "+str(p))
(t, p) = scipy.stats.ttest_ind(lunch_st_read, lunch_red_read)
print("P-Value for the Reading score: "+str(p))
(t, p) = scipy.stats.ttest_ind(lunch_st_write, lunch_red_write)
print("P-Value for the Writing score: "+str(p))
```

```python
# Choosing the rows where test preparation course type is "none", creating subsets for all exams separately
test_none_rows = data_all["test preparation course"] == "none"
test_none_math = data_all[test_none_rows]['math score']
test_none_read = data_all[test_none_rows]['reading score']
test_none_write = data_all[test_none_rows]['writing score']

# Choosing the rows where test preparation course type is "completed", creating subsets for all exams separately
test_comp_rows = data_all["test preparation course"] == "completed" 
test_comp_math = data_all[test_comp_rows]['math score']
test_comp_read = data_all[test_comp_rows]['reading score']
test_comp_write = data_all[test_comp_rows]['writing score']

# Calculate the statistical significance for each exam based on completion of test preparation course
(t, p) = scipy.stats.ttest_ind(test_none_math, test_comp_math)
print("P-Value for the Math score: "+str(p))
(t, p) = scipy.stats.ttest_ind(test_none_read, test_comp_read)
print("P-Value for the Reading score: "+str(p))
(t, p) = scipy.stats.ttest_ind(test_none_write, test_comp_write)
print("P-Value for the Writing score: "+str(p))
```
